# Installing Required R Packages

To install the R packages required for this project, follow these instructions. It has to be done only one time.

*NB: These instructions assume you are using SAS Viya Workbench, but you can adjust them in any supported R environment.*

In VS Code, if prompted to select a kernel, choose **Jupyter Kernel...** and select the **Recommended** R environment.

Run the following cell:

```r
packages <- readLines("R_Requirements.txt")
install.packages(packages)
```
